```python
# Compute the correlation matrix
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(7, 7))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1.0, center=0,vmin=-1.0,
            square=True, linewidths=0.5, cbar_kws={"shrink": .5})
```

```python
from sklearn.model_selection import train_test_split

'''
X = df.iloc[:, :-1]
y = df.iloc[:,-1]
'''


#'''
X=df[['loc', 'v(g)', 'ev(g)', 'iv(g)', 'n', 'v', 'l', 'd', 'i', 'e', 'b', 't','lOCode', 'lOComment', 'lOBlank', 
       'locCodeAndComment', 'uniq_Op','uniq_Opnd', 'total_Op', 'total_Opnd', 'branchCount']]
y=df['defects']
#'''

'''
list=['defects']
X = df.drop(list,axis=1)
y = df['defects']
'''

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,random_state=42)


'''
X=np.array(X)
X.shape
y=np.array(y)
y.shape
'''
```

```python
from sklearn.neural_network import MLPClassifier as mlp
```

```python
clf=mlp()
clf.fit(X_train,y_train)
predictions=clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predictions))
print("--------------------")
print(classification_report(y_test,predictions))
```

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train,y_train)
predictionsvm = svc_model.predict(X_test)
```

```python
print(confusion_matrix(y_test,predictionsvm))
print("--------------------")
print(classification_report(y_test,predictionsvm))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(X_train,y_train)
predictionsdtree=dtree.predict(X_test)
```

```python
print(confusion_matrix(y_test,predictionsdtree))
print("--------------------")
print(classification_report(y_test,predictionsdtree))
```

```python
from sklearn.ensemble import RandomForestClassifier
rtree=RandomForestClassifier(n_estimators=200)
rtree.fit(X_train,y_train)
rpredictions=rtree.predict(X_test)
```

```python
print(confusion_matrix(y_test,rpredictions))
print("--------------------")
print(classification_report(y_test,rpredictions))
```
